```python
import xarray as xr
import glob
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
fpath = glob.glob('data/BACND*oxygen*nc')  # search for patterns
ds = [xr.open_dataset(f) for f in fpath]
```

```python
a = []
for i in range(10):
    a.append(i)
```

```python

fig = plt.figure()
ax  = fig.add_subplot(1,1,1, projection=ccrs.PlateCarree())


aux = ds[0]
lon = aux.attrs['lon']
lat = aux.attrs['lat']
ax.scatter(lon, lat)


aux = ds[1]
lon = aux.attrs['lon']
lat = aux.attrs['lat']
ax.scatter(lon, lat, s=5)

aux = ds[2]
lon = aux.attrs['lon']
lat = aux.attrs['lat']
ax.scatter(lon, lat, s=1)


ax.coastlines()
ax.set_extent([-130.5, -123.08, 47.6, 50.15])
```

```python
fig = plt.figure()
ax  = fig.add_subplot(1,1,1, projection=ccrs.PlateCarree())

for i in range(len(ds)):
    aux = ds[i]
    lon = aux.attrs['lon']
    lat = aux.attrs['lat']
    ax.scatter(lon, lat)
    print(i,lon,lat)

ax.coastlines()
ax.set_extent([-130.5, -123.08, 47.6, 50.15])
```
